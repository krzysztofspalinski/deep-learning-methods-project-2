```python
import pandas as pd

arch1 = pd.read_csv("./results/architecture-1-results.csv")
arch2 = pd.read_csv("./results/architecture-2-results.csv")
arch3 = pd.read_csv("./results/architecture-3-results.csv")
arch4 = pd.read_csv("./results/architecture-4-results.csv")
results = {"Architecture 1": arch1,
           "Architecture 2": arch2,
           "Architecture 3": arch3,
           "Architecture 4": arch4}
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
sns.palplot(sns.color_palette("Set2"))
sns.set_palette("dark")
```

```python
plots = {"Accuracy": 'accuracy',
        "Loss": 'loss',
        "Validation Accuracy": 'val_accuracy',
        "Validation Loss": 'val_loss'}
lims = False
```

```python

for k, v in plots.items():
    
    plt.figure(figsize=(15, 9))
    
    for name, architecture in results.items():
        plt.plot([i+1 for i in range(len(architecture[v + '_mean']))],
                 architecture[v + '_mean'],
                 label=name)

    #plt.title('czna suma grantw w ramach konkursw NCN')
    plt.legend(loc='upper left')
    plt.title(k)
    plt.subplots_adjust(bottom=0.2)
    plt.savefig('./results/resnet_experiment' + v + '.png', bbox_inches='tight')
    plt.show()
```

```python

for k, v in plots.items():
    
    plt.figure(figsize=(15, 9))
    
    for name, architecture in results.items():
        plt.errorbar([i+1 for i in range(len(architecture[v + '_mean']))],
                 architecture[v + '_mean'], yerr= architecture[v + '_std'], uplims=True, lolims=True,
                 label=name)

    #plt.title('czna suma grantw w ramach konkursw NCN')
    plt.legend(loc='upper left')
    plt.title(k)
    plt.subplots_adjust(bottom=0.2)
    plt.savefig('./results/resnet_experiment' + v + '_errorbars.png', bbox_inches='tight')
    plt.show()
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict

def get_image(img: np.ndarray):
    """
    input: img should be a row of cifar-100 dataset.
    returns: numpy 3D array, to be used with plt.imshow() function.
    """
    return np.reshape(img, (32, 32, 3), order='F').transpose(1, 0, 2)
```

```python
raw_data = unpickle('../data/data_batch_1')
```

```python
X_train = raw_data[b'data']
```

```python
y_train = np.array(raw_data[b'labels'])
```

```python
label_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python


fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3)
fig.set_figheight(6.5)
fig.set_figwidth(9)


img_num = 0
ax1.set_title(label_names[y_train[img_num]])
ax1.imshow(get_image(X_train[img_num]))

img_num = 1
ax2.set_title(label_names[y_train[img_num]])
ax2.imshow(get_image(X_train[img_num]))

img_num = 8
ax3.set_title(label_names[y_train[img_num]])
ax3.imshow(get_image(X_train[img_num]))

img_num = 3
ax4.set_title(label_names[y_train[img_num]])
ax4.imshow(get_image(X_train[img_num]))

img_num = 4
ax5.set_title(label_names[y_train[img_num]])
ax5.imshow(get_image(X_train[img_num]))

img_num = 6
ax6.set_title(label_names[y_train[img_num]])
ax6.imshow(get_image(X_train[img_num]))

plt.savefig('data.png', bbox_inches='tight')
```
